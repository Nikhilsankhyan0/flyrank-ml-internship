## 2. Target or proxy

## Target / Proxy

The target used in the starter dataset is a proxy label called **is_declining_label**, which is created from the `trend_direction` column.

A webpage classified as **down** is treated as a page that may require content review.

In a more advanced version of the project, the target could be future content decline or future recovery based on later performance windows.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

The primary success metric is **Precision@50**, which measures how many of the top 50 recommended webpages actually belong to the target class.

This metric matches the real business workflow because SEO teams usually review only a limited number of pages at a time.

Additional metrics such as Precision@20 and Average Precision may also be useful.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why Machine Learning Beats a Fixed Rule

A fixed rule relies on only a few manually selected conditions and may miss complex relationships between different features.

Machine learning can learn patterns from multiple signals such as impressions, clicks, CTR, average position, content age, freshness, engagement rate, and traffic trends simultaneously.

This allows the system to produce a more reliable and data-driven ranking for content review while still supporting human decision-making.
